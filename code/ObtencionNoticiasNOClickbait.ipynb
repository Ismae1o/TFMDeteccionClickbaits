```python
from newspaper import Article, build
import pandas as pd
import time
import random

# Fuentes fiables (sin El Confidencial)
trusted_sources = {
    "El Pa√≠s": "https://elpais.com",
    "La Vanguardia": "https://www.lavanguardia.com",
    "eldiario.es": "https://www.eldiario.es",
    "Cinco D√≠as": "https://cincodias.elpais.com"
}

# Datos finales
articles_data = []
articles_per_source = 100

# Conjuntos para evitar duplicados
seen_urls = set()
seen_titles = set()

for name, url in trusted_sources.items():
    print(f"\nüì° Procesando {name} ({url})...")
    paper = build(url, memoize_articles=False)
    articles = paper.articles
    collected = 0

    for article in articles:
        if collected >= articles_per_source:
            break
        try:
            article.download()
            article.parse()

            title = article.title.strip()
            content = article.text.strip()
            url = article.url

            # Comprobar duplicados
            if (title and content and len(content) > 300 and
                url not in seen_urls and title not in seen_titles):

                articles_data.append({
                    "title": title,
                    "content": content,
                    "url": url,
                    "clickbait": 0
                })
                collected += 1
                seen_urls.add(url)
                seen_titles.add(title)

                print(f"  [{collected}/100] {title[:60]}...")

            time.sleep(random.uniform(0.5, 1.1))

        except Exception:
            continue

print(f"\n‚úÖ Total de art√≠culos recopilados: {len(articles_data)}")

# Guardar a CSV
df = pd.DataFrame(articles_data)
df.to_csv("noticias_no_clickbait.csv", index=False, encoding="utf-8")
print("üìÅ Archivo guardado como 'noticias_no_clickbait.csv'")
```
